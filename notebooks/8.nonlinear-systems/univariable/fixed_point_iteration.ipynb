```python
import numpy as np
import matplotlib.pyplot as plt
```

## $f(x)=cos(x)-x$, find its root

```python
f = lambda x: np.cos(x) - x
```

## Fixed Point Iteration

### From Scratch

```python
def fixed_point(f, x0, epsilon=1e-10, num_iter=10000):
    g = lambda x: f(x) + x
    x = []
    x.append(x0)
    for k in range(num_iter):
        x.append(g(x[k]))
        if abs(g(x[k+1]) - g(x[k])) < abs(x[k+1] - x[k]):
            if abs(x[k+1] - x[k]) < epsilon:
                print("The number of iteration: ", k+1)
                return x[-1], x
        else:
            print("The result is divergent!")
            return None, None
```

```python
root, root_n = fixed_point(f, 0)
root
```

```python
g = lambda x: f(x) + x

plt.figure(figsize=(8, 8))
xn = np.linspace(-1, 1, 100)
plt.plot(xn, g(xn), label="$g(x)$")
plt.scatter(root_n, g(root_n), c='g', label="iteration root")
plt.scatter(root, root, c='r', marker='*', s=100, label="root")
plt.plot(xn, xn, linestyle='--', alpha=0.5, label="$y=x$")
plt.legend(fontsize=14)
plt.show()
```

### Using `Scipy`

```python
import scipy.optimize as optimize

g = lambda x: f(x) + x
optimize.fixed_point(g, -100)
```

```python
f1 = lambda x: np.exp(x) + np.exp(-x) - 5 - x
root1, _ = fixed_point(f1, 3)
root1
```
