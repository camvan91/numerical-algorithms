```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 60
np.random.seed(123)
x = np.random.uniform(-1, 7, n)
f = lambda x: np.cos(x) + 2
y = f(x) +  np.random.randn(n) * 0.3
```

```python
interpolant = interp1d(x, y, kind="nearest")
xn = np.linspace(x.min(), x.max(), 100)
plt.figure(figsize=(10, 8))
plt.plot(x, y, "ro", label="Nodes")
plt.plot(xn, f(xn), "b", label="$y=\sin(3x)$")
plt.plot(xn, interpolant(xn), "g--", label="Interpolation")
plt.title("Nearest-neighbor approximation", fontsize=14)
plt.legend(loc=4)
plt.show()
```
