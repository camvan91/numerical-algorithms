### Piecewise Linear Basis (Hat Function)

```python
nodes = np.linspace(0, 5, 6)
```

```python
def psi(x, X ):
    diff = np.abs(X - x)
    idx = int(np.where(diff == diff.min())[0])
    if idx > 0 and idx < len(X) - 1 :
        #print(idx)
        if x - X[idx] <= 0:
            y = (x - X[idx-1]) / (X[idx] - X[idx-1])
        else:
            y = (X[idx+1] - x) / (X[idx+1] - X[idx])
    else:
        y = 0
    return y
```

```python
nodes = np.linspace(0, 4, 5)
xn = np.linspace(0, 5, 1000)

yn = np.zeros(n)
for i,x in enumerate(xn):
    yn[i] = psi(x, nodes)

plt.figure(figsize=(10, 8))
plt.plot(nodes, np.zeros(len(nodes)), 'ko')    
colors = ['b', 'g', 'r', 'orange']
for i in range(1, len(nodes)-1):
    l = (nodes[i-1] + nodes[i]) / 2
    r = (nodes[i] + nodes[i+1]) / 2
    xni = xn[(xn > l) & (xn < r)]
    yni = yn[(xn > l) & (xn < r)]
    plt.plot(xni, yni, c=colors[i-1], label="$\psi_{}$".format(i))

plt.legend(fontsize=14)
plt.ylim(0, 2)
plt.show()
```
